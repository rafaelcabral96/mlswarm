The goal here is to find the minimum of several univariate and multivariate function.

For that it is necessary to:
1. func = function(lambda x: ...) - create a function object 
2. func = init_cloud(...) - Define array of initial particle positions
3. func.minimize(...) - Define the algorithms parameters and start the algorithm

```python
from mlswarm import function
import numpy as np
```

## f(x) = x^2 - (cos(4 Pi x) -1)

```python
#we want to minimize function f
func = function(lambda x: x[0]**2 - (np.cos(4 * np.pi *x[0]) - 1))

#starting cloud: 50 points spread in the interval [2,3]
func.init_cloud(np.array([np.arange(50)/50 + 2]))

#minimize func
func.minimize(max_iterations = 500, 
              var_epsilon = 0, 
              learning_rate = 0.01,
              method = "swarm_derivfree",
              kernel_a = 0.01,
              alpha_init = 0,
              alpha_rate = 1,
              beta=0,
              gamma=1,
              verbose = False)

#plot cost history
func.plot_cost_history()
```

## f(x) = x^4 - 2*(cos(2*Pi*x) - 1) - x^2

```python
#another function
func = function(lambda x: x[0]**4 - 2*(np.cos(2*np.pi*x[0]) - 1) - x[0]**2)

func.init_cloud(np.array([np.arange(50)/50 + 2]))

func.minimize(max_iterations = 100, 
              var_epsilon = 0, 
              learning_rate = 0.01,
              method = "swarm_derivfree",
              kernel_a = 0.01,
              alpha_init = 0,
              alpha_rate = 10,
              beta=0,
              gamma=1,
              verbose = False)

func.plot_cost_history()
```

## f(x,y,z) = x^2 + y^2 + z^2

```python
func = function(lambda x: x[0]**2 + x[1]**2 + x[2]**2)

func.init_cloud(np.array([np.arange(50)/50 + 2]*3))

func.minimize(max_iterations = 100, 
              var_epsilon = 0, 
              learning_rate = 0.01,
              method = "swarm_derivfree",
              kernel_a = 0.01,
              alpha_init = 0,
              alpha_rate = 1,
              beta=0,
              gamma=1,
              verbose = False)

func.plot_cost_history()
```

## Rosenbrock function considering N_dim dimensions. Minimum at f(1,....1) = 0

```python
N_dim=10
func = function(lambda x: np.mean([(x[i+1]-x[i]**2)**2 + (1-x[i])**2 for i in range(N_dim-1)],axis=0))

func.init_cloud(np.array([np.arange(50)/50 + 10]*N_dim))

func.minimize(max_iterations = 1000, 
              var_epsilon = 0, 
              learning_rate = 0.001,
              method = "swarm_derivfree",
              kernel_a = 0.01,
              alpha_init = 0,
              alpha_rate = 10,
              beta=0,
              gamma=1,
              verbose = False)

func.plot_cost_history()
```
