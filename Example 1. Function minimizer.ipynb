The goal here is to find the minimum of several univariate and multivariate function.

For that it is necessary to:
1. func = function(lambda x: ...) - create a function object 
2. func = init_cloud(...) - Define array of initial particle positions
3. func.minimize(...) - Define the algorithms parameters and start the algorithm

```python
from mlswarm.mlswarm import function
import numpy as np
```

```python
#we want to minimize function f
func1 = function(lambda x: x[0]**2 - (np.cos(4 * np.pi *x[0]) - 1), "f(x) = x^2 - (cos(4 Pi x) -1)")

#starting cloud: 50 points with normal(0,2) distribution
func1.init_cloud(np.array([np.arange(50)/50 + 2]))

#minimize func
func1.minimize(max_iterations = 2000 , var_epsilon = 0, 
              learning_rate = 0.01,
              method = "swarm_derivfree",
              kernel_a = 0.01, alpha_init = 0, alpha_rate = 1, beta=0, gamma=1,
              verbose = False, track_history = True)

#plot cost history
func1.plot_everything("Example 1", f_dim = 1)
```

```python
#plot cost history
from mlswarm.mlswarm import integral_on_gaussian_measure_util

integral_on_gaussian_measure_util(func1.func, func1.cloud_history_mean, func1.cloud_var_history)
```

```python
func2 = function(lambda x: x[0]**4 - 2*(np.cos(2*np.pi*x[0]) - 1) - x[0]**2, "f(x) = x^4 - 2*(cos(2*Pi*x) - 1) - x^2")

func2.init_cloud(np.array([np.arange(50)/50 + 2]))

func2.minimize(max_iterations = 500, var_epsilon = 0, 
              learning_rate = 0.01,
              method = "swarm_derivfree",
              kernel_a = 0.1, alpha_init = 0, alpha_rate = 1, beta=0, gamma=1,
              verbose = False)

func2.plot_cost_history()
```

```python
N_dim = 5
f = lambda x: 10*N_dim + np.sum([x[i]**2 - 10*np.cos(2*np.pi*x[i])  for i in range(N_dim)],axis=0)
func3 = function(f, "Rastrigin function")

func3.init_cloud(np.array(np.random.normal(2,1,(N_dim,100))))

func3.minimize(max_iterations = 5000, var_epsilon = 0, 
              learning_rate = 0.001,
              method = "swarm_derivfree",
              kernel_a = 0.1, alpha_init = 1, alpha_rate = 0.01, beta=0, gamma=1,
              verbose = False)

func3.plot_cost_history()
```
