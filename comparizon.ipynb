```python
from mlswarm.mlswarm import function
import numpy as np
np.random.seed(42)
track_history = False
```

```python
f = lambda x: np.take(x,0,axis=0)**2
name = "Quadratic Function"
func = function(f, name)

func.init_cloud([[-1],[-0.5],[0.2],[2]])

func.minimize({'kernel_a':1,
               'learning_rate':1,
               'var_epsilon': 0,
               'max_iterations': 200})

data_swarm_euler_adaptive_1 = np.repeat(func.cost_history_mean,func.N)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Datasets/func1.csv", header = None).T
data = data.applymap(f)

plt.figure()
data.plot()
plt.plot(data_swarm_euler_adaptive_1)
#plt.plot(data_swarm_euler_adaptive_2)
plt.yscale('log')
plt.xlabel("Function Evaluations")
plt.ylabel("Function Value")
plt.xlim(0,100)
plt.legend(["DifferentialEvolution", "MeshSearch", "NelderMead", "SimulatedAnnealing","Swarm"],bbox_to_anchor=(1.04,1),loc="upper left")
```
