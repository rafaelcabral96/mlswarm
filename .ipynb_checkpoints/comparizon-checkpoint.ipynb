```python
from mlswarm.mlswarm import function
import numpy as np
np.random.seed(42)
track_history = False
```

```python
f = lambda x: np.take(x,0,axis=0)**2
name = "Quadratic Function"
func = function(f, name)

func.init_cloud([[-1],[-0.5],[0.2],[2]])

func.minimize({'kernel_a':1,
               'learning_rate':1,
               'var_epsilon': 0,
               'max_iterations': 200})

data_swarm_euler_adaptive_1 = np.repeat(func.cost_history_mean,func.N)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Datasets/func1.csv", header = None).T
data = data.applymap(f)

plt.figure()
data.plot()
plt.plot(data_swarm_euler_adaptive_1)
#plt.plot(data_swarm_euler_adaptive_2)
plt.yscale('log')
plt.xlabel("Function Evaluations")
plt.ylabel("Function Value")
plt.xlim(0,100)
plt.legend(["DifferentialEvolution", "MeshSearch", "NelderMead", "SimulatedAnnealing","Swarm"],bbox_to_anchor=(1.04,1),loc="upper left")
```

```python
f = lambda x: np.take(x,0, axis = 0)**2 - (np.cos(4 * np.pi *np.take(x,0, axis = 0)) - 1)
name = "pol_and_cosine"
func = function(f, name)

func.init_cloud(np.linspace(-1.5,2,5)[:,np.newaxis])

func.minimize({'kernel_a':1,
               'learning_rate':1,
               'var_epsilon': 0,
               'max_iterations': 200})

data_swarm_euler_adaptive_1 = np.repeat(func.cost_history_mean,func.N)
```

```python
f = lambda x: np.sin(10*np.pi*np.take(x,0,axis=0))/(2*np.take(x,0,axis=0)) + (np.take(x,0,axis=0)-1)**4
name = "Gramacy & Lee Function" 
func = function(f, name)

func.init_cloud(np.linspace(-1,1,6)[:,np.newaxis])

func.minimize({'var_epsilon': 0.01,
               'max_iterations': 1000})

data_swarm_euler_adaptive_1 = np.repeat(func.cost_history_mean,func.N)
```

```python
f = lambda x: (1.5 - np.take(x,0,axis=0) + np.take(x,0,axis=0)*np.take(x,1,axis=0))**2 + (2.25 - np.take(x,0,axis=0) + np.take(x,0,axis=0)*np.take(x,1,axis=0)**2)**2 + (2.625 - np.take(x,0,axis=0) + np.take(x,0,axis=0)*np.take(x,1,axis=0)**3)**2
func = function(f, "Beale function")

x=np.linspace(-1, 2, 5)
y=np.linspace(-1, 2, 5)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'var_epsilon': 0,
               'alpha_rate':1,
               'max_iterations': 100})

data_swarm_euler_adaptive_1 = np.repeat(func.cost_history_mean,func.N)
```

```python
f = lambda x: -20* np.exp(-0.2*np.sqrt(0.5*(np.take(x,0,axis=0)**2 + np.take(x,1,axis=0)**2))) - np.exp(0.5*(np.cos(2*np.pi*np.take(x,0,axis=0))+np.cos(2*np.pi*np.take(x,1,axis=0)))) + np.exp(1) + 20
func = function(f,"Ackley function")

x=np.linspace(-1, 2, 3)
y=np.linspace(-1, 2, 3)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'var_epsilon': 0,
               'alpha_rate':1,
               'max_iterations': 100})

data_swarm_euler_adaptive_1 = np.repeat(func.cost_history_mean,func.N)
```
