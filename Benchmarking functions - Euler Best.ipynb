```python
from mlswarm.mlswarm import function
import numpy as np
np.random.seed(42)

func_list = {}
track_history = False
```

## 1D Functions

```python
f = lambda x: np.take(x,0,axis=0)**2
func = function(f, "Quadratic Function")

func.init_cloud([[-1],[-0.5],[0.2],[2]])

func.minimize({'algorithm':'euler',
               'learning_rate':0.1})

func_list[func.name] = func
func.plot_everything(folder_name = func.name, f_dim = 1, log = True)
```

```python
f = lambda x: np.take(x,0, axis = 0)**2 - (np.cos(4 * np.pi *np.take(x,0, axis = 0)) - 1)
func = function(f, "pol_and_cosine")

func.init_cloud(np.linspace(-2,1,5)[:,np.newaxis])

func.minimize({'algorithm':'euler'})

func_list[func.name] = func
func.plot_everything(folder_name = func.name, f_dim = 1, log = True)
```

```python
f = lambda x: np.take(x,0,axis=0)**4 - 2*(np.cos(2*np.pi*np.take(x,0,axis=0)) -1) - np.take(x,0,axis=0)**2
func = function(f, "pol_and_cosine_2")

func.init_cloud(np.linspace(-2,1.5,5)[:,np.newaxis])

func.minimize({'algorithm':'euler'})

func_list[func.name] = func
func.plot_everything(folder_name = func.name, f_dim = 1, log = True, limits = [-2,2])
```

```python
f = lambda x: np.sin(10*np.pi*np.take(x,0,axis=0))/(2*np.take(x,0,axis=0)) + (np.take(x,0,axis=0)-1)**4
name = "Gramacy & Lee Function" 
func = function(f, name)

func.init_cloud(np.linspace(-1,1,6)[:,np.newaxis])

func.minimize({'algorithm':'euler',
               'learning_rate': 0.001, 
               'var_epsilon':0.01})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 1, log = True, limits = [-2,2])
```

## 2D Functions

```python
f = lambda x: (1.5 - np.take(x,0,axis=0) + np.take(x,0,axis=0)*np.take(x,1,axis=0))**2 + (2.25 - np.take(x,0,axis=0) + np.take(x,0,axis=0)*np.take(x,1,axis=0)**2)**2 + (2.625 - np.take(x,0,axis=0) + np.take(x,0,axis=0)*np.take(x,1,axis=0)**3)**2
func = function(f, "Beale function")

x=np.linspace(-1, 2, 5)
y=np.linspace(-1, 2, 5)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'algorithm':'euler',
               'var_epsilon': 0.01,
               'learning_rate': 0.001})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -20* np.exp(-0.2*np.sqrt(0.5*(np.take(x,0,axis=0)**2 + np.take(x,1,axis=0)**2))) - np.exp(0.5*(np.cos(2*np.pi*np.take(x,0,axis=0))+np.cos(2*np.pi*np.take(x,1,axis=0)))) + np.exp(1) + 20
func = function(f,"Ackley function")

x=np.linspace(-1, 2, 3)
y=np.linspace(-1, 2, 3)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'algorithm':'euler'})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: 0.5 + (np.sin(np.take(x,0,axis=0)**2 - np.take(x,1,axis=0)**2)**2 - 0.5)/(1 +0.01*(np.take(x,0,axis=0)**2 + np.take(x,1,axis=0)**2)**2)
func = function(f,"Schaffer function N. 2")

x=np.linspace(-1, 2, 3)
y=np.linspace(-1, 2, 3)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'algorithm':'euler'})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: np.sin(3*np.pi*x[0])**2 + (x[0] - 1)**2 * (1 + np.sin(3*np.pi*x[1])) + (x[1]-1)**2*(1+np.sin(2*np.pi*x[1])**2)
func = function(f,"Lvi function N.13")

x=np.linspace(-1, 2, 3)
y=np.linspace(-1, 2, 3)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'algorithm':'euler'})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -0.0001*np.power(np.abs(np.sin(x[0])*np.sin(x[1])*np.exp(np.abs(100 - np.sqrt(x[0]**2 + x[1]**2)/np.pi)))+1,0.1)
func = function(f, "Cross-in-Tray Function")

x=np.linspace(-1, 2, 3)
y=np.linspace(-1, 2, 3)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'algorithm':'euler',
               'var_epsilon':0.01})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -(np.sin(x[0]-x[1])**2*np.sin(x[0]+x[1])**2)/(np.sqrt(x[0]**2 + x[1]**2))
func = function(f, "Keane Function")

x=np.linspace(-1, 2, 3)
y=np.linspace(-1, 2, 3)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T
func.init_cloud(coords)

func.minimize({'algorithm': 'euler',
               'var_epsilon': 0.01})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -np.cos(x[0])*np.cos(x[1])*np.exp(-(x[0]-np.pi)**2 - (x[1]-np.pi)**2)
func = function(f, "Easom Function")

x=np.linspace(-2, 6, 3)
y=np.linspace(-2, 6, 3)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T

func.init_cloud(coords)

func.minimize({'algorithm': 'euler',
               'var_epsilon': 0.01})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True,limits =[0,np.pi*2,0,np.pi*2])
```

# >2D Function

```python
N_dim=2
f = lambda x: np.sum([100*(x[i+1]-x[i]**2)**2 + (1-x[i])**2 for i in range(N_dim-1)],axis=0)
func = function(f, "Rosenbrock function 3D")

#np.random.seed(42)
x=np.linspace(-1, 2, 5)
y=np.linspace(-1, 2, 5)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel())).T

func.init_cloud(coords)

func.minimize({'algorithm': 'euler',
               'learning_rate': 0.0001,
               'var_epsilon': 0.01})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
N_dim=5
f = lambda x: (np.sin(x[0])**2+ np.sin(x[1])**2 - np.exp(-x[0]**2 - x[1]**2))*np.exp(-np.sin(np.sqrt(np.abs(x[0])))**2- np.sin(np.sqrt(np.abs(x[1])))**2)
func = function(f, "Xin-She Yang N. 4 Function")

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(1,2,(30,N_dim))))

func.minimize({'algorithm': 'euler',
               'learning_rate': 0.1,
               'var_epsilon': 0.01})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 5, log = True)
```

```python
N_dim = 3
f = lambda x: np.sum([(x[i]**2 - i)**2 for i in range(N_dim)],axis=0)
func = function(f, "Qing Function")

func.init_cloud(np.array(np.random.normal(0,3,(20, N_dim))))

func.minimize({'algorithm': 'euler',
               'learning_rate': 0.01,
               'var_epsilon': 0.001})

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 3, log = True,limits =[-3,3,-3,3])
```
