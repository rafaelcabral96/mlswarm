```python
from mlswarm.mlswarm import function
import numpy as np
np.random.seed(42)

func_list = {}
track_history = False
```

## 1D Functions

```python
f = lambda x: x[0]**2
name = "Quadratic Function"
func = function(f, name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(3,2,(1,50))))

func.minimize(max_iterations = 10000, var_epsilon = 0.0001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.001, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=10, gamma_rate = 0,
              verbose = False, track_history = track_history)

func.cloud_var
func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 1, log = True)
```

```python
f = lambda x: x[0]**2 - (np.cos(4 * np.pi *x[0]) - 1)
name = "pol_and_cosine"
func = function(f, name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(3,1,(1,50))))

func.minimize(max_iterations = 5000, var_epsilon = 0.00001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.001, alpha_init = 0, alpha_rate = 10, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 1, log = True)
```

```python
f = lambda x: x[0]**4 - 2*(np.cos(2*np.pi*x[0]) -1) - x[0]**2
name = "pol_and_cosine_2"
func = function(f, name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(3,1,(1,50))))

func.minimize(max_iterations = 10000, var_epsilon = 0.0001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.001, alpha_init = 0, alpha_rate = 10, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 1, log = True, limits = [-2,2])
```

```python
f = lambda x: np.sin(10*np.pi*x[0])/(2*x[0]) + (x[0]-1)**4
name = "Gramacy & Lee Function" 
func = function(f, name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(-3,1,(1,50))))

func.minimize(max_iterations = 100000, var_epsilon = 0.0001, 
              learning_rate = 0.001,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.01, alpha_init = 0, alpha_rate = 10, beta=0,  gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 1, log = True, limits = [-2,2])
```

## 2D Functions

```python
f = lambda x: (1.5 - x[0] + x[0]*x[1])**2 + (2.25 - x[0] + x[0]*x[1]**2)**2 + (2.625 - x[0] + x[0]*x[1]**3)**2
name = "Beale function"
func = function(f, name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(-2,1,(2,50))))

func.minimize(max_iterations = 10000, var_epsilon = 0.0001, 
              learning_rate = 0.001,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.1, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -20* np.exp(-0.2*np.sqrt(0.5*(x[0]**2 + x[1]**2))) - np.exp(0.5*(np.cos(2*np.pi*x[0])+np.cos(2*np.pi*x[1]))) + np.exp(1) + 20
name = "Ackley function"
func = function(f,name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(3,1,(2,100))))

func.minimize(max_iterations = 100000, var_epsilon = 0.0001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.001, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: 0.5 + (np.sin(x[0]**2 - x[1]**2)**2 - 0.5)/(1 +0.01*(x[0]**2 + x[1]**2)**2)
name = "Schaffer function N. 2"
func = function(f,name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(0,1,(2,20))))

func.minimize(max_iterations = 100000, var_epsilon = 0.00001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.001, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: np.sin(3*np.pi*x[0])**2 + (x[0] - 1)**2 * (1 + np.sin(3*np.pi*x[1])) + (x[1]-1)**2*(1+np.sin(2*np.pi*x[1])**2)
name = "Lévi function N.13"
func = function(f,name)

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(3,1,(2,50))))

func.minimize(max_iterations = 100000, var_epsilon = 0.00001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.01, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -0.0001*np.power(np.abs(np.sin(x[0])*np.sin(x[1])*np.exp(np.abs(100 - np.sqrt(x[0]**2 + x[1]**2)/np.pi)))+1,0.1)
func = function(f, "Cross-in-Tray Function")

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(3,2,(2,50))))

func.minimize(max_iterations = 100000, var_epsilon = 0.0001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.1, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -(np.sin(x[0]-x[1])**2*np.sin(x[0]+x[1])**2)/(np.sqrt(x[0]**2 + x[1]**2))
func = function(f, "Keane Function")

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(2,3,(2,50))))

func.minimize(max_iterations = 100000, var_epsilon = 0.0001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.01, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate=0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
f = lambda x: -np.cos(x[0])*np.cos(x[1])*np.exp(-(x[0]-np.pi)**2 - (x[1]-np.pi)**2)
func = function(f, "Easom Function")

#2D grid in [0,5]x[0,5] -> interesting -> for some function 2D grid is better than gaussian initializations
x=np.linspace(-2, 6, 10)
y=np.linspace(-2, 6, 10)
xx,yy=np.meshgrid(x,y)
coords=np.array((xx.ravel(), yy.ravel()))

func.init_cloud(coords)
#func.init_cloud(False, auto = True, search_space = np.array([[-5,-5],[5,5]]), N = 50, function_dim =2)    
    
func.minimize(max_iterations = 10000, var_epsilon = 0.0001, 
              learning_rate = 1,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.01, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True,limits =[0,np.pi*2,0,np.pi*2])
```

```python
N_dim=2
f = lambda x: (np.sin(x[0])**2+ np.sin(x[1])**2 - np.exp(-x[0]**2 - x[1]**2))*np.exp(-np.sin(np.sqrt(np.abs(x[0])))**2- np.sin(np.sqrt(np.abs(x[1])))**2)
func = function(f, "Xin-She Yang N. 4 Function")

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(1,2,(N_dim,200))))

func.minimize(max_iterations = 100000, var_epsilon = 0.0001, 
              learning_rate = 0.01, N_sample = False,
              method = "swarm_derivfree",
              kernel_a = 0.001, alpha_init = 0, alpha_rate = 1, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 2, log = True)
```

```python
N_dim = 10
f = lambda x: np.sum([(x[i]**2 - i)**2 for i in range(N_dim)],axis=0)
func = function(f, "Qing Function")

np.random.seed(42)
func.init_cloud(np.array(np.random.normal(0,1,(N_dim,100))))

func.minimize(max_iterations = 100000, var_epsilon = 0.0001, 
              learning_rate = 0.01,
              method = "swarm_derivfree", N_sample = False,
              kernel_a = 0.0001, alpha_init = 0, alpha_rate = 10, beta=0, gamma_init=1, gamma_rate = 0,
              verbose = False, track_history = track_history)

func_list[func.name] = func
#func.plot_everything(folder_name = func.name, f_dim = 10, log = True,limits =[-3,3,-3,3])
```
